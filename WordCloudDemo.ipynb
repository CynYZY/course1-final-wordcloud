```python
filename = "Demo.txt"
with open(filename, encoding='UTF-8') as f:
    mytext = f.read()
```

```python
print(mytext)
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate(mytext)
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
